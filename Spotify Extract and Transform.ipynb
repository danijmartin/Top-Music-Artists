```python
#dependencies
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials #To access authorised Spotify data
from config import client_id, client_secret
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import time
import requests
```

```python
file= 'Resources/artists.csv'
# file= '../etl-group-project/artists.csv'
```

```python
artists_df = pd.read_csv(file)
#print(artists_df.columns)
```

```python
new_artists_df = artists_df.drop(columns=["mbid", "artist_lastfm", "country_lastfm", "tags_mb","tags_lastfm", "scrobbles_lastfm", "ambiguous_artist"]).head(50)
new_artists_df.head()
```

```python
# gettings artists into an array to make an API/scape call later
artist_array = new_artists_df.loc[:,"artist_mb"]
print(artist_array)
```

```python
# Setting up Splinter 
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# Connecting spotify search website
url = 'https://open.spotify.com/search'
browser.visit(url)

artist_ids = []

for artist in artist_array:
    browser.find_by_tag('input').fill(artist)
    time.sleep(3)
    artist_url =browser.find_by_tag("a[class='f7ebc3d96230ee12a84a9b0b4b81bb8f-scss']")['href']
    artist_ids.append(artist_url)
    print(artist_url)

browser.quit()

artist_ids
```

```python
print(len(artist_ids))

confirm = set(artist_ids)
print(len(confirm))
```

```python
ids_df = pd.DataFrame(artist_ids)
ids_df.head()
```

```python
ids_df = ids_df[0].str.split('/', expand=True)
ids_df.head()
```

```python
new_artists_df['spotify_id'] = ids_df[4]
new_artists_df.head()
```

```python
new_artists_df.to_csv('Resources/updated_artists.csv', index=False)
```

```python
spotify_artist_id = new_artists_df['spotify_id']
spotify_artist_id
```

```python
spotify_artist_name = new_artists_df['artist_mb']
spotify_artist_name
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id,
                                                           client_secret))
pop_song = []

for spotid in spotify_artist_id:    
    track = {}
    results = sp.artist_top_tracks(spotid)
    
    track['artist'] = results['tracks'][0]['artists'][0]['name'] #artist name
    track['name'] = results['tracks'][0]['name'] #track name
    track['popularity'] = results['tracks'][0]['popularity'] #popularity rating

    pop_song.append(track)
    
print(pop_song)
```

```python
pop_song_df = pd.DataFrame(pop_song)
pop_song_df
```

```python
pop_song_df.to_csv('Resources/pop_song.csv', index=False)
```
